```python
import os
import pandas as pd
import numpy as np
def convert_npy_to_csv(folder_path):
    # Get a list of all files in the specified folder
    files = os.listdir(folder_path)

    # Iterate through each file in the folder
    for file_name in files:
        # Check if the file has a .mat extension
        if file_name.endswith('.npy'):
            # Construct the full path for the .mat file
            npy_file_path = os.path.join(folder_path, file_name)

            # Load data from the .npy file
            npy_data = np.load(npy_file_path)
            #print(npy_data)
            
            # converting each .npy file to .csv
            # accessing the last key of a dictionary which is the form of array
            #last_key = list(npy_data.keys())[-1]
            
            # accessing the array of the last key
            # converting the array in to python standard array form which is dataframe
            df=pd.DataFrame(npy_data)
            print(df)

            # assigning aname for each .csv to be created by removing the last four characters from the original name 
            # and adding .csv 
            file_name = file_name[:-4]+'.csv'
            
            # locating the output folde
            output_file_path = os.path.join(folder_path, file_name)
            
            #writing the array to csv file 
            df.to_csv(output_file_path, index=False)
            #df.to_csv(file_name)
            
folder_path = 'C:/Users/Ewunetie/OneDrive - University of Gondar,/2022-2023 files/Courses/Project/data/Online data/4 ECG Lead 2 Dataset PhysioNet (Open Access)/db_npy/svdb_npy'
convert_npy_to_csv(folder_path)
```
