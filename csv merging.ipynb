### Merging or concatinating CSV
this will be used for csv files only relatively simillar size and allignment and out of memory error will pop up for files having one is column array and another is row array with each length of 1250

```python
import pandas as pd
import os
import xlrd # module for reading and writting excel files
# creating empty dataframe for intializing and storing the value after each iteration
dfs = pd.DataFrame()
path = 'D:/ML Project/Data/mit-bih-long-term-ecg-database-1.0.0/New folder'
files = os.listdir(path)
for file_name in files:
    # Check if the file has a .mat extension
    if file_name.endswith('.csv'):
        # Construct the full path for the .mat file
# path = 'C:/Users/Ewunetie/OneDrive - University of Gondar,/2022-2023 files/Courses/ML Project/Data/Preprocessed'
        csv_file_path = os.path.join(path, file_name)        
        # Load data from the .csv file
        
        csv_data = pd.read_csv(csv_file_path)
        #print(csv_data)
        # merging 
        dfs=pd.concat([dfs, csv_data],axis=0, join='outer')
#print(dfs) 
df=pd.DataFrame(dfs)
df = df.dropna()
output_file_path = os.path.join(path,"Merged.csv")
df.to_csv(output_file_path, index=False)
df
```
