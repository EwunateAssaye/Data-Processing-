```python
import pandas as pd
import os
#import xlrd # module for reading and writting excel files
# creating empty dataframe for intializing and storing the value after each iteration
folder_path = 'C:/Users/Ewunetie/OneDrive - University of Gondar,/2022-2023 files/Courses/ML Project/Data/Online data/3 MIT-BIH Arrhythmia Database (Simple CSVs)/row'
output_path = 'C:/Users/Ewunetie/OneDrive - University of Gondar,/2022-2023 files/Courses/ML Project/Data/Online data/3 MIT-BIH Arrhythmia Database (Simple CSVs)/Lead II'
files = os.listdir(folder_path)
for file_name in files:
    # Check if the file has a .mat extension
    if file_name.endswith('.csv'):
        # Construct the full path for the .mat file
        csv_file_path = os.path.join(folder_path, file_name)        
        # Load data from the .csv file
        csv_data = pd.read_csv(csv_file_path)
        
        # Selecting the specific column with the column header electrocardiogram signal (lead 2)
        df1=csv_data['MLII']
                
        df=pd.DataFrame(df1)
        
        file_name = file_name[:-4]+'Lead II.csv'
        output_file_path = os.path.join(output_path,file_name)
        df.to_csv(output_file_path, index=False)
        
```
